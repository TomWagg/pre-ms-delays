```python
import delay
import cogsworth
import numpy as np
import astropy.units as u
```

```python
p = cogsworth.pop.Population(100, final_kstar1=[14])
```

```python
p.sample_initial_binaries()
p.sample_initial_galaxy()

p.initial_galaxy._Z = np.repeat(0.02, len(p))
p.initial_galaxy._tau = np.repeat(10, len(p)) * u.Gyr
p._initial_binaries["metallicity"] = p._initial_galaxy.Z
p._initial_binaries["tphysf"] = p._initial_galaxy.tau.to(u.Myr).value

p.perform_stellar_evolution()
```

```python
delayed_initC, int_bins = delay.delay_companions(p.initC, delay.get_delays(p.initC),
                                                 pool=None, BSEdict=p.BSE_settings)
```

```python
q = p[:]
```

```python
q._initC = delayed_initC
```

```python
q.perform_stellar_evolution()
```
